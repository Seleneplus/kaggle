####  <span style="color:red">**Question 1.1**</span> **Activation function plays an important role in modern Deep NNs. For each of the activation functions below, state its output range, find its derivative (show your steps), and plot the activation fuction and its derivative**

<span style="color:red">**(a)**</span> Exponential linear unit (ELU): $\text{ELU}(x)=\begin{cases}
0.1\left(\exp(x)-1\right) & \text{if}\,x\leq0\\
x & \text{if}\,x>0
\end{cases}$
<div style="text-align: right"><span style="color:red">[1.5 points]</span></div>

<span style="color:red">**(b)**</span> Gaussian Error Linear Unit (GELU): $\text{GELU}(x)=x\Phi(x)$ where $\Phi(x)$ is the `probability cummulative function` of the standard Gaussian distribution or $\Phi(x) = \mathbb{P}\left(X\leq x\right)$ where $X \sim N\left(0,1\right)$. In addition, the GELU activation fuction (the link for the [main paper](https://arxiv.org/pdf/1606.08415v5.pdf)) has been widely used in the state-of-the-art Vision for Transformers (e.g., here is the link for [the main ViT paper](https://arxiv.org/pdf/2010.11929v2.pdf)).
<div style="text-align: right"><span style="color:red">[1.5 points]</span></div>

(a) for x > 0, the output is x, so the range is (0,+∞),derivative is 1.
    for x <=0, the output is 0.1*(exp(x) - 1), when x close to negative infinity, the output approaches -0.1. thus, the overall output range of ELU is [−0.1,+∞), derivative is 0.1*exp(x).

(b)the output range is all real numbers $\mathbb{R}$, because x is a real number and $\Phi(x)$ is a probalibity that ranges from 0 to 1. The derivative of the GELU function is: $\Phi(x)$ + $x \cdot \phi(x)$, where $x \cdot \phi(x)$ is the probability density function (PDF) of the standard normal distribution, defined as: $x \cdot \phi(x)$ = $\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)
$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define ELU function and its derivative
def elu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=0.1):
    return np.where(x > 0, 1, alpha * np.exp(x))

# Define GELU function and its derivative
def gelu(x):
    return x * norm.cdf(x)

def gelu_derivative(x):
    return norm.cdf(x) + x * norm.pdf(x)

# Create x values
x = np.linspace(-3, 3, 400)

# Compute y values for ELU and GELU functions and their derivatives
elu_y = elu(x)
elu_y_deriv = elu_derivative(x)
gelu_y = gelu(x)
gelu_y_deriv = gelu_derivative(x)

# Plot ELU function and its derivative
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, elu_y, label='ELU(x)', color='orange')
plt.plot(x, elu_y_deriv, label="ELU'(x)", linestyle='--', color='red')
plt.title('ELU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Plot GELU function and its derivative
plt.subplot(1, 2, 2)
plt.plot(x, gelu_y, label='GELU(x)', color='orange')
plt.plot(x, gelu_y_deriv, label="GELU'(x)", linestyle='--', color='red')
plt.title('GELU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Custom loss class combining cross-entropy and entropy
class CustomLoss(nn.Module):
    def __init__(self, lambda_entropy=0.1):
        super(CustomLoss, self).__init__()
        self.lambda_entropy = lambda_entropy
        self.cross_entropy = nn.CrossEntropyLoss()

    def forward(self, outputs, targets):
        ce_loss = self.cross_entropy(outputs, targets)
        p = torch.softmax(outputs, dim=1)
        entropy = -torch.sum(p * torch.log(p + 1e-9), dim=1).mean()
        total_loss = ce_loss + self.lambda_entropy * entropy
        return total_loss

# Simple neural network with two hidden layers
class CustomNN(nn.Module):
    def __init__(self, n1, n2, activation):
        super(CustomNN, self).__init__()
        self.fc1 = nn.Linear(784, n1)
        self.fc2 = nn.Linear(n1, n2)
        self.fc3 = nn.Linear(n2, 10)
        self.activation = activation

    def forward(self, x):
        x = x.view(-1, 784)  # Flatten input
        x = self.activation(self.fc1(x))
        x = self.activation(self.fc2(x))
        x = self.fc3(x)
        return x

# Initialize model, optimizer, and custom loss function
model = CustomNN(n1=40, n2=40, activation=nn.ReLU())
optimizer = optim.RMSprop(model.parameters(), lr=0.0001)
criterion = CustomLoss(lambda_entropy=0.1)

# Function to calculate accuracy
def calculate_accuracy(outputs, targets):
    _, predicted = torch.max(outputs, 1)  # Get the index of the max log-probability
    correct = (predicted == targets).sum().item()
    return correct

# Training loop
for epoch in range(30):
    model.train()
    total_loss = 0
    correct_predictions = 0
    total_samples = 0

    for data, target in train_loader:
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        correct_predictions += calculate_accuracy(output, target)
        total_samples += target.size(0)  # Batch size

    accuracy = correct_predictions / total_samples
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(train_loader):.4f}, Accuracy: {accuracy * 100:.2f}%")
```

Sharpness-aware minimization (SAM) (i.e., [link for main paper](https://openreview.net/pdf?id=6Tm1mposlrM) from Google Deepmind) is a simple yet but efficient technique to improve the generalization ability of deep learning models on unseen data examples. In your research or your work, you might potentially use this idea. Your task is to read the paper and implement *Sharpness-aware minimization (SAM)*. Finally, you need to apply SAM to the best architecture found in **Question 2.3**.   

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

# Define the SAM optimizer
class SAM(torch.optim.Optimizer):
    def __init__(self, params, base_optimizer, rho=0.05, adaptive=False, **kwargs):
        assert rho >= 0.0, f"Invalid rho, should be non-negative: {rho}"

        # Convert params to a list to avoid the generator issue
        params = list(params)

        defaults = dict(rho=rho, adaptive=adaptive, **kwargs)
        super(SAM, self).__init__(params, defaults)

        self.base_optimizer = base_optimizer(self.param_groups, **kwargs)
        self.param_groups = self.base_optimizer.param_groups

    def step(self, closure=None):
        if closure is None:
            raise ValueError("SAM requires closure, but it was not provided")

        # Perform first step
        loss = closure()
        loss.backward()

        # Compute gradient norm
        grad_norm = torch.norm(torch.stack([p.grad.norm() for p in self.param_groups[0]['params'] if p.grad is not None]))

        # First step: perturb the weights
        for group in self.param_groups:
            for p in group['params']:
                if p.grad is None:
                    continue
                e_w = p.grad * self.defaults['rho'] / grad_norm
                p.data.add_(e_w)

        # Zero the gradients, second step
        self.base_optimizer.zero_grad()
        loss = closure()
        loss.backward()
        self.base_optimizer.step()

        # Restore original weights
        for group in self.param_groups:
            for p in group['params']:
                if p.grad is None:
                    continue
                e_w = p.grad * self.defaults['rho'] / grad_norm
                p.data.sub_(e_w)

# Define a simple neural network
class CustomNN(nn.Module):
    def __init__(self, n1, n2, activation):
        super(CustomNN, self).__init__()
        self.fc1 = nn.Linear(784, n1)
        self.fc2 = nn.Linear(n1, n2)
        self.fc3 = nn.Linear(n2, 10)
        self.activation = activation

    def forward(self, x):
        x = x.view(-1, 784)  # Flatten the input
        x = self.activation(self.fc1(x))
        x = self.activation(self.fc2(x))
        x = self.fc3(x)
        return x

# Accuracy calculation function
def calculate_accuracy(model, data_loader, device):
    model.eval()  # Set the model to evaluation mode
    correct = 0
    total = 0
    with torch.no_grad():
        for data, target in data_loader:
            data, target = data.to(device), target.to(device)
            outputs = model(data)
            _, predicted = torch.max(outputs, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()
    return correct / total

# Initialize values for the model
best_n1 = 40  # First hidden layer size
best_n2 = 40  # Second hidden layer size
best_activation = nn.ReLU()  # Activation function

# Set up device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create the model and optimizer
model = CustomNN(n1=best_n1, n2=best_n2, activation=best_activation).to(device)

# Set up base optimizer
base_optimizer = optim.Adam  # Use Adam optimizer as base

# Create SAM optimizer
optimizer = SAM(model.parameters(), base_optimizer, rho=0.05, lr=0.001)

# Loss function
criterion = nn.CrossEntropyLoss()

# Load the dataset (example using MNIST)
transform = transforms.Compose([transforms.ToTensor()])
train_loader = DataLoader(datasets.MNIST('./data', train=True, download=True, transform=transform), batch_size=64, shuffle=True)
test_loader = DataLoader(datasets.MNIST('./data', train=False, transform=transform), batch_size=1000, shuffle=False)

# Training loop
for epoch in range(10):
    model.train()
    running_loss = 0.0
    for data, target in train_loader:
        data, target = data.to(device), target.to(device)

        def closure():
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            return loss

        # Call closure explicitly to get loss
        loss = closure()
        optimizer.step(closure)
        running_loss += loss.item()  # Accumulate loss

    # Calculate accuracy and print loss and accuracy
    train_loss = running_loss / len(train_loader)
    test_accuracy = calculate_accuracy(model, test_loader, device)

    print(f'Epoch {epoch+1}, Loss: {train_loss:.4f}, Accuracy: {test_accuracy * 100:.2f}%')
```

**First, you need to implement the block of our CNN in the class `YourBlock`. You can ignore `use_skip` and `skip connection` for simplicity. However, you cannot earn full marks for this question.**

```python
class YourBlock(nn.Module):
    def __init__(self, in_feature_maps, out_feature_maps, drop_rate=0.2, batch_norm=True, use_skip=True):
        super(YourBlock, self).__init__()
        self.use_skip = use_skip
        # Convolution layer
        self.conv = nn.Conv2d(in_feature_maps, out_feature_maps, kernel_size=3, stride=1, padding=1)

        # Batch Normalization layer (if enabled)
        self.batch_norm = nn.BatchNorm2d(out_feature_maps) if batch_norm else None

        # Activation function
        self.activation = nn.GELU()
        # Max Pooling layer
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)

        # Dropout layer
        self.dropout = nn.Dropout(drop_rate) if drop_rate > 0 else None

        # Skip connection (only if use_skip is True)
        if self.use_skip:
            self.skip_conv = nn.Conv2d(in_feature_maps, out_feature_maps, kernel_size=1, stride=2, padding=0)
        else:
            self.skip_conv = None

    def forward(self, x):
        residual = x  # Keep input as residual

        x = self.conv(x)

        if self.batch_norm is not None:
            x = self.batch_norm(x)

        x = self.activation(x)
        x = self.pool(x)

        if self.dropout is not None:
            x = self.dropout(x)

        # If skip connection is used, adjust the residual with skip_conv
        if self.use_skip:
            residual = self.skip_conv(residual)
        else:
            # If skip connection is not used, we still need to match the dimensions
            residual = F.avg_pool2d(residual, kernel_size=2, stride=2)

            # If the number of channels doesn't match, pad with zeros
            if residual.size(1) != x.size(1):
                zero_pad = torch.zeros(x.size(0), x.size(1) - residual.size(1), x.size(2), x.size(3), device=x.device)
                residual = torch.cat([residual, zero_pad], dim=1)

        # Add the residual (skip connection) to the output
        x = x + residual

        return x

# Example usage
in_channels = 3
out_channels = 16
batch_size = 4
input_size = 32

# Create an instance of YourBlock
block = YourBlock(in_channels, out_channels)

# Create a random input tensor
input_tensor = torch.randn(batch_size, in_channels, input_size, input_size)

# Pass the input through the block
output = block(input_tensor)

```

```python
class YourCNN(nn.Module):
    def __init__(self, list_feature_maps=[16, 32, 64], drop_rate=0.2, batch_norm=True, use_skip=True, num_classes=20):
        super(YourCNN, self).__init__()
        layers = []

        in_features = 3  # Initial number of input channels (for example, 3 for RGB images)
        for out_features in list_feature_maps:
            layers.append(YourBlock(in_feature_maps=in_features, out_feature_maps=out_features,
                                    drop_rate=drop_rate, batch_norm=batch_norm, use_skip=use_skip))
            in_features = out_features

        # Adaptive pooling to reduce dimensionality
        layers.append(nn.AdaptiveAvgPool2d((1, 1)))
        layers.append(nn.Flatten())

        # Fully connected layer to output the final class predictions
        self.fc = nn.Linear(in_features, num_classes)
        self.block = nn.Sequential(*layers)

    def forward(self, x):
        x = self.block(x)
        x = self.fc(x)
        return x


```

```python
#Your code here
import torch
import numpy as np

# Define the mixup function
def mixup_data(x, y, alpha=1.0):
    '''Returns mixed inputs, pairs of targets, and lambda'''
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1

    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(x.device)

    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]

    return mixed_x, y_a, y_b, lam

# Define a loss function for mixup
def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# Training function with Mixup
def fit_with_mixup(model, train_loader, val_loader, optimizer, num_epochs=20, alpha=0.2):
    model.to(device)
    criterion = torch.nn.CrossEntropyLoss()

    for epoch in range(num_epochs):
        model.train()
        for i, (inputs, targets) in enumerate(train_loader):
            inputs, targets = inputs.to(device), targets.to(device)

            # Apply mixup
            inputs, targets_a, targets_b, lam = mixup_data(inputs, targets, alpha)

            # Forward pass
            outputs = model(inputs)
            loss = mixup_criterion(criterion, outputs, targets_a, targets_b, lam)

            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Validate the model
        model.eval()
        val_loss = 0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = model(inputs)
                val_loss += criterion(outputs, targets).item()
                _, predicted = torch.max(outputs, 1)
                total += targets.size(0)
                correct += (predicted == targets).sum().item()

        # Calculate validation accuracy
        val_acc = 100 * correct / total
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {val_loss/len(val_loader):.4f}, Validation Accuracy: {val_acc:.2f}%')

# Initialize your model and optimizer
model = YourCNN(list_feature_maps=[32, 64, 128], drop_rate=0.1, batch_norm=True, use_skip=True, num_classes=20)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Train the model with Mixup
fit_with_mixup(model, train_loader, val_loader, optimizer, num_epochs=10, alpha=0.4)

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingLR
import numpy as np

# Define the CutMix function
def cutmix_data(x, y, alpha=1.0):
    '''Returns cutmixed inputs, pairs of targets, and lambda'''
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1

    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(x.device)

    # Generate a random box
    W, H = x.size(2), x.size(3)
    cut_rat = np.sqrt(1. - lam)
    cut_w = int(W * cut_rat)
    cut_h = int(H * cut_rat)

    # Uniformly sample the center of the box
    cx = np.random.randint(W)
    cy = np.random.randint(H)

    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)

    # Create the cutmix image
    x[:, :, bbx1:bbx2, bby1:bby2] = x[index, :, bbx1:bbx2, bby1:bby2]

    y_a, y_b = y, y[index]

    # Adjust lambda based on the area of the cutout region
    lam = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (W * H))

    return x, y_a, y_b, lam

# Define a loss function for CutMix
def cutmix_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# Improved training function with CutMix
def fit_with_cutmix(model, train_loader, val_loader, optimizer, scheduler, num_epochs=50, alpha=1.0):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    criterion = nn.CrossEntropyLoss()

    best_val_acc = 0
    patience = 10
    patience_counter = 0
    training_results = []  # Save results during training for analysis

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        correct = 0
        total = 0

        for i, (inputs, targets) in enumerate(train_loader):
            inputs, targets = inputs.to(device), targets.to(device)

            # Apply CutMix
            inputs, targets_a, targets_b, lam = cutmix_data(inputs, targets, alpha)

            # Forward pass
            outputs = model(inputs)
            loss = cutmix_criterion(criterion, outputs, targets_a, targets_b, lam)

            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # Gradient clipping
            optimizer.step()

            train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += targets.size(0)
            correct += (lam * predicted.eq(targets_a.data).cpu().sum().float()
                        + (1 - lam) * predicted.eq(targets_b.data).cpu().sum().float())

        train_acc = 100 * correct / total
        train_loss = train_loss / len(train_loader)

        # Validate the model
        model.eval()
        val_loss = 0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = model(inputs)
                val_loss += criterion(outputs, targets).item()
                _, predicted = torch.max(outputs, 1)
                total += targets.size(0)
                correct += (predicted == targets).sum().item()

        val_loss = val_loss / len(val_loader)
        val_acc = 100 * correct / total

        # Learning rate scheduling
        scheduler.step()

        # Save training and validation results for later use (e.g. plotting loss curves and analysis)
        training_results.append({
            "epoch": epoch + 1,
            "train_loss": train_loss,
            "train_acc": train_acc.item(),
            "val_loss": val_loss,
            "val_acc": val_acc
        })

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, '
              f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')

        # Early stopping
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f'Early stopping triggered after {epoch+1} epochs')
                break

    print(f'Best validation accuracy: {best_val_acc:.2f}%')
    return model, training_results

# Define your CNN model (assuming you have a YourCNN class defined)
class YourCNN(nn.Module):
    def __init__(self, list_feature_maps, drop_rate, batch_norm, use_skip, num_classes):
        super(YourCNN, self).__init__()
        self.features = nn.ModuleList()
        in_channels = 3  # Assuming RGB input
        for out_channels in list_feature_maps:
            self.features.append(YourBlock(in_channels, out_channels, drop_rate, batch_norm, use_skip))
            in_channels = out_channels

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(list_feature_maps[-1], num_classes)

    def forward(self, x):
        for block in self.features:
            x = block(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.fc(x)
        return x

# Initialize your model and optimizer
model = YourCNN(list_feature_maps=[64, 128, 256, 512], drop_rate=0.3, batch_norm=True, use_skip=True, num_classes=20)
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=50)  # Cosine annealing scheduler

# Train the model with CutMix
trained_model, training_results = fit_with_cutmix(model, train_loader, val_loader, optimizer, scheduler, num_epochs=20, alpha=1.0)

# Print the training results for analysis
print(training_results)
```

<span style="color:red">**Question 3.5:**</span> Implement the **one-versus-all (OVA)** loss. The details are as follows:
- You need to apply `the sigmoid activation function` to logits $h = [h_1, h_2,...,h_M]$ instead of `the softmax activation` function as usual to obtain $p = [p_1, p_2,...,p_M]$, meaning that $p_i = sigmoid(h_i), i=1,...,M$. Note that $M$ is the number of classes.
- Given a data example $x$ with the ground-truth label $y$, the idea is to maximize the likelihood $p_y$ and to minimize the likelihoods $p_i, i \neq y$. Therefore, the objective function is to find the model parameters to
  - $\max\left\{ \log p_{y}+\sum_{i\neq y}\log(1-p_{i})\right\}$ or equivalently $\min\left\{ -\log p_{y}-\sum_{i\neq y}\log(1-p_{i})\right\}$.
  - For example, if $M=3$ and $y=2$, you need to minimize $\min\left\{ -\log(1-p_{1})-\log p_{2}-\log(1-p_{3})\right\}$.

```python
#Your code here
import torch
import torch.nn as nn
import torch.optim as optim
import os

# PGD Attack implementation for adversarial training
def pgd_attack(model, images, labels, eps=0.0313, alpha=0.002, iters=10):
    images = images.clone().detach().to(device)
    labels = labels.to(device)
    original_images = images.clone().detach()

    for _ in range(iters):
        images.requires_grad = True
        outputs = model(images)

        model.zero_grad()
        cost = nn.CrossEntropyLoss()(outputs, labels)
        cost.backward()

        adv_images = images + alpha * images.grad.sign()
        eta = torch.clamp(adv_images - original_images, min=-eps, max=eps)
        images = torch.clamp(original_images + eta, min=0, max=1).detach()

    return images

# FGSM Attack implementation for evaluation
def fgsm_attack(model, images, labels, eps=0.0313):
    images = images.clone().detach().to(device)
    labels = labels.to(device)
    images.requires_grad = True
    outputs = model(images)

    loss = nn.CrossEntropyLoss()(outputs, labels)
    model.zero_grad()
    loss.backward()

    adv_images = images + eps * images.grad.sign()
    adv_images = torch.clamp(adv_images, min=0, max=1).detach()

    return adv_images

# Adversarial training function using PGD
def adversarial_train(model, train_loader, val_loader, optimizer, num_epochs=50, eps=0.0313, alpha=0.002, iters=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    best_val_acc = 0

    for epoch in range(num_epochs):
        model.train()
        for i, (inputs, targets) in enumerate(train_loader):
            inputs, targets = inputs.to(device), targets.to(device)

            # Generate adversarial examples using PGD
            adv_inputs = pgd_attack(model, inputs, targets, eps, alpha, iters)

            # Forward pass with adversarial examples
            optimizer.zero_grad()
            outputs = model(adv_inputs)
            loss = nn.CrossEntropyLoss()(outputs, targets)
            loss.backward()
            optimizer.step()

        # Validation step
        val_acc = evaluate(model, val_loader)
        print(f"Epoch [{epoch+1}/{num_epochs}], Validation Accuracy: {val_acc:.2f}%")

        # Save the best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            if not os.path.exists('./models'):
                os.makedirs('./models')
            torch.save(model.state_dict(), './models/best_robust_model.pth')

# Evaluation function for adversarial robustness
def evaluate_robustness(model, test_loader, attack_type="pgd", eps=0.0313, alpha=0.002, iters=20):
    model.eval()
    correct = 0
    total = 0
    attack_fn = pgd_attack if attack_type == "pgd" else fgsm_attack

    for inputs, targets in test_loader:
        inputs, targets = inputs.to(device), targets.to(device)

        # Apply adversarial attack
        adv_inputs = attack_fn(model, inputs, targets, eps=eps, alpha=alpha, iters=iters)

        outputs = model(adv_inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += targets.size(0)
        correct += (predicted == targets).sum().item()

    return 100 * correct / total

# Define your CNN model (assuming you have a YourCNN class defined)
class YourCNN(nn.Module):
    def __init__(self, list_feature_maps, drop_rate, batch_norm, use_skip, num_classes):
        super(YourCNN, self).__init__()
        self.features = nn.ModuleList()
        in_channels = 3  # Assuming RGB input
        for out_channels in list_feature_maps:
            self.features.append(YourBlock(in_channels, out_channels, drop_rate, batch_norm, use_skip))
            in_channels = out_channels

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(list_feature_maps[-1], num_classes)

    def forward(self, x):
        for block in self.features:
            x = block(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.fc(x)
        return x

# Initialize your model and optimizer
model = YourCNN(list_feature_maps=[64, 128, 256, 512], drop_rate=0.3, batch_norm=True, use_skip=True, num_classes=20)
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)

# Train the model with adversarial training
adversarial_train(model, train_loader, val_loader, optimizer, num_epochs=50, eps=0.0313, alpha=0.002, iters=10)

# Load the best robust model
model.load_state_dict(torch.load('./models/best_robust_model.pth'))

# Evaluate the model on PGD and FGSM attacks
pgd_robust_acc = evaluate_robustness(model, test_loader, attack_type="pgd", eps=0.0313, alpha=0.002, iters=20)
fgsm_robust_acc = evaluate_robustness(model, test_loader, attack_type="fgsm", eps=0.0313)

print(f"Robust Accuracy against PGD: {pgd_robust_acc:.2f}%")
print(f"Robust Accuracy against FGSM: {fgsm_robust_acc:.2f}%")


```
